```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# image of a playing card
image = cv2.imread("card.JPG", cv2.IMREAD_GRAYSCALE)

# plot the image
plt.figure(figsize=(5, 5))
plt.imshow(image, cmap="gray")
plt.show()
```

```python
_, binary_image = cv2.threshold(image, 100, 255, cv2.THRESH_BINARY)

# plot the binary image
plt.figure(figsize=(5, 5))
plt.imshow(binary_image, cmap="gray")
plt.show()
```

```python
# apply sobel kernel to the image
sobel_x = cv2.Sobel(binary_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(binary_image, cv2.CV_64F, 0, 1, ksize=3)

# plot the sobel x image
plt.figure(figsize=(5, 5))
plt.imshow(sobel_x, cmap="gray")
plt.show()

# plot the sobel y image
plt.figure(figsize=(5, 5))
plt.imshow(sobel_y, cmap="gray")
plt.show()

# calculate the magnitude of the gradient
magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

# plot the magnitude of the gradient
plt.figure(figsize=(5, 5))
plt.imshow(magnitude, cmap="gray")
plt.show()

# calculate the direction of the gradient
direction = np.arctan2(sobel_y, sobel_x)

# plot the direction of the gradient
plt.figure(figsize=(5, 5))
plt.imshow(direction, cmap="gray")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
N = 64  # Length of the signal
t = np.arange(N)  # Time indices

# Signal with frequency components at +16 and -16, and a DC offset
dc_offset = 1
frequency = 16 / N
x = np.cos(2 * np.pi * frequency * t) + dc_offset

# Compute the FFT and shift it
X = np.fft.fft(x)
X_shifted = np.fft.fftshift(X)

# Frequency bins
freqs = np.fft.fftfreq(N, d=1)
freqs_shifted = np.fft.fftshift(freqs)
freq_bins = freqs_shifted * N  # Scale to frequency bins

# Plot the shifted Fourier spectrum
plt.stem(freq_bins, np.abs(X_shifted))
plt.xlabel('Frequency Bin')
plt.ylabel('Magnitude')
plt.title('Shifted Fourier Spectrum')
plt.grid(True)
plt.show()
```
